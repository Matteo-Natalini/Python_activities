```python
import pandas as pd
```

```python
# Read the data sets

movies = pd.read_excel('Data/movies_merge.xlsx')

ott =  pd.read_csv('Data/ott_merge.csv')
```

```python
# Validate whether the data is loaded correctly

print(movies.shape)
movies.head()
```

```python
print(ott.shape)
ott.head()
```

```python
# Describe the data types of the individual columns

print(movies.dtypes)
print(ott.dtypes)
```

```python
# Combine the two DataFrames

movott_merge = movies.merge(ott, how='left', on='ID')

print(movott_merge.shape)
movott_merge.head()
```

```python
movott_concat = pd.concat([movies, ott], axis=0)

print(movott_concat.shape)
movott_concat.head()
```
