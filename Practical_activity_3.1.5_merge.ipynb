```python
import pandas as pd
```

```python
# Read the data sets

movies = pd.read_excel('Data/movies_merge.xlsx')

ott =  pd.read_csv('Data/ott_merge.csv')
```

```python
# Validate whether the data is loaded correctly

print(movies.shape)
movies.head()
```

```python
print(ott.shape)
ott.head()
```

```python
# Describe the data types of the individual columns

print(movies.dtypes)
print(ott.dtypes)
```

```python
# Combine the two DataFrames

movott = movies.merge(ott, how='left', on='ID')

print(movott.shape)
movott.head()
```

```python
movott_concat = pd.concat([movies, ott], axis=0)

print(movott_concat.shape)
movott_concat.head()
```

## Question 1: How many films from each year (released from 2012 to the present) were watched on Netflix?

```python
mo_gby = movott[movott['Year'] >=2012].groupby('Year')[['Netflix']]

mo_gby.sum()#.reset_index()
```

## Question 2: What is the average runtime of movies released each year?

```python
mo_gby1 = movott[movott['Year'] >=2012].groupby('Year')[['Runtime']]

mo_gby1.mean()
```

## Question 3: What are the best and worst reviews that movies received on Rotten Tomatoes?

```python
mo_gby2 = movott[movott['Year'] >=2012].groupby('Year')[['Rotten Tomatoes']]

mo_gby2.agg(['min', 'max'])
```
